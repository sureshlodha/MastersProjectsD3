```python
   
from pandas import Series, DataFrame
from statsmodels.tsa.stattools import adfuller
from sklearn.cross_validation import train_test_split 
from datetime import datetime
from itertools import count
from statsmodels.tsa.arima_model import ARIMA
from math import sqrt
from sklearn.metrics import mean_squared_error
from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.arima_model import ARIMAResults
import matplotlib.dates as mdates
import datetime
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import time as time
import warnings

warnings.filterwarnings("ignore")   



def complete_data_analysis(complete_data):
    #month year wise analysis
    complete_data = complete_data[complete_data.Year != 2017].sort_values('Year')
    grouped_for_total_analysis = complete_data.groupby(['Year'],as_index=False)['Total_Gallons'].agg('sum')
    grouped_for_total_analysis = grouped_for_total_analysis.sort_values('Year')
    grouped_for_total_analysis_by_month = complete_data.groupby(['Year_Month'],as_index=False)['Total_Gallons'].agg('sum')
    print(grouped_for_total_analysis)

    #year analysis
    pyplot.bar(grouped_for_total_analysis['Year'], grouped_for_total_analysis['Total_Gallons'])
    pyplot.xlabel('Date', fontsize=14, color='brown')
    pyplot.ylabel('Total_Gallons in Millions', fontsize=14, color='brown')
    pyplot.title("Complete_Data_Analysis_Bar")
    pyplot.savefig("Complete_Data_Analysis_Bar.png")
    pyplot.show()
    
     #year analysis
    x = grouped_for_total_analysis_by_month['Year_Month'].apply(lambda x:x/100)
    pyplot.plot(x, grouped_for_total_analysis_by_month['Total_Gallons'])
    #grouped_for_year_analysis.bar()
    pyplot.xlabel('Date', fontsize=14, color='brown')
    pyplot.title("Complete_Data_Analysis_Time_Series")
    pyplot.ylabel('Total_Gallons in Millions', fontsize=14, color='brown')
    pyplot.savefig("Complete_Data_Analysis_Time_Series.png")
    pyplot.show()
    return

def plotByPostalCode(complete_data):
    #loop for each postal and customer class
    grouped_postal_code = complete_data.groupby(['Postal_Code'])
    #for each Postal_Code
    for postal_code,df_postal_code in grouped_postal_code:
        title = ('PostalCode%s' % (postal_code))
        df_date_analysis = df_postal_code.groupby(['Year_Month','Date'],as_index=False)['Total_Gallons'].agg('sum')
        call_predict(df_date_analysis,title)
    return

def plotByCustomerClass(complete_data, mapping):
    #loop for each postal and customer class
    grouped_customer_class = complete_data.groupby(['Customer_Class'])
    for customer_class, df_customer_class in grouped_customer_class:
        title = ('Usage Customer Class %s') % mapping[customer_class]
        df_date_analysis = df_customer_class.groupby(['Year_Month','Date'], as_index=False)['Total_Gallons'].agg('sum')
        call_predict(df_date_analysis,title)
    return

# evaluate an ARIMA model for a given order (p,d,q)
def evaluate_arima_model(X, arima_order):
    warnings.filterwarnings("ignore")
	# prepare training dataset
    train_size = int(len(X) * 0.66)
    train, test = X[0:train_size], X[train_size:]
    history = [x for x in train]
    # make predictions
    predictions = list()
    try:
        for t in range(len(test)):
            model = ARIMA(history, order=arima_order)
            model_fit = model.fit(disp=0)
            yhat = model_fit.forecast()[0]
            predictions.append(yhat)
            history.append(test[t])
    except:
        return 111111111111111
	# calculate out of sample error
    mse = mean_squared_error(test, predictions)
    error = sqrt(mse)
    return error
 
# evaluate combinations of p, d and q values for an ARIMA model
def evaluate_models(dataset, p_values, d_values, q_values):
    warnings.filterwarnings("ignore")
    partition = (int)(len(dataset)*0.66)
    dataset = dataset[0:partition]
    dataset = dataset.astype('float32')
    best_score, best_cfg = float("inf"), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    mse = evaluate_arima_model(dataset, order)
                    if mse < best_score:
                        best_score, best_cfg = mse, order
                        print('ARIMA%s MSE=%.3f' % (order,mse))
                except:
                        continue;
    print('Best ARIMA%s MSE=%.3f' % (best_cfg, best_score))
    return best_cfg


def call_predict(data,title):
    title = title.strip()
    title = title.replace(" ", "_")
    data = data.sort_values('Year_Month')
    warnings.filterwarnings("ignore")
    
    # evaluate parameters
    p_values = [0, 1, 2]
    d_values = range(0,2)
    q_values = range(0,2)
    warnings.filterwarnings("ignore")
    order = evaluate_models(data['Total_Gallons'].values, p_values, d_values, q_values)
    if(order == (0,0,0)):
        print('Data is white noise, cannot predict')
        print("\n")
    else:
        prediction(data,order,title)
    return

def prediction(data, order, title):
    warnings.filterwarnings("ignore")
    data = data.sort_values('Year_Month')
    
    partition = int(len(data)*0.75)
    dataset = data['Total_Gallons'].values
    train = dataset[0:partition]
    test = dataset[partition:]
    
    # walk-forward validation
    history = [x for x in train]
    predictions = list()
    
    try:
        for i in range(len(test)):
            # predict
            model = ARIMA(history, order=(order))
            model_fit = model.fit(trend='nc',disp=-1)
            yhat = model_fit.forecast()[0]
            predictions.append(yhat)
            
            # observation
            obs = test[i]
            history.append(obs)
            print('>Predicted=%.3f, Expected=%3.f' % (yhat, obs))

    except:
        print("exception1")
        try:
            for i in range(len(test)):
                model = ARIMA(history, order=(order))
                model_fit = model.fit(disp=0)
                yhat = model_fit.forecast()[0]
                predictions.append(yhat)

                # observation
                obs = test[i]
                history.append(obs)
                print('>Predicted=%.3f, Expected=%3.f' % (yhat, obs))
            
        except:
                print("exception2")
                return
     
    spacing = 2
    dates = data['Date'][partition:]
    fig, ax = plt.subplots()
    ax.xaxis_date()
    print(title)
    years = mdates.YearLocator()   # every year
    months = mdates.MonthLocator()  # every month
    yearsFmt = mdates.DateFormatter('%Y')
    ax.autoscale_view()
    yearsFmt = mdates.DateFormatter('%Y')
    ax.xaxis.set_major_locator(years)
    ax.xaxis.set_major_formatter(yearsFmt)
    ax.xaxis.set_minor_locator(months)
    ax.plot(test,'r-',label="test_data")
    ax.plot(predictions,'b-',label='predicted_values')
    ax.set_title(title)
    ax.legend(loc=4, fontsize=12)
    ax.set_xticks(np.arange(len(dates)))
    ax.set_xticklabels(dates)
    ax.set_ylabel('Total Gallons (in Millions)', fontsize = 12)
    ax.set_xlabel('Date', fontsize = 12)
    fig.autofmt_xdate()
    for label in ax.xaxis.get_ticklabels()[::spacing]:
        label.set_visible(False)
    fig.set_size_inches(4.5, 4.5)
    title = ('%s' % (title +'.png'))
    print("Title",title)
    pyplot.savefig(title)
    pyplot.show()
    return

            
#function to
def convert_string_data_to_int(df):
    columns = df.columns.values
    mapping = {}

    #for each column
    for column in columns:
        text_to_int_dict = {}

        def convert_to_int(val):
            return text_to_int_dict[val]

        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0

            for unique in unique_elements:
                if unique not in text_to_int_dict:
                    text_to_int_dict[unique] = x
                    mapping[x] = unique
                    x+=1
            
            df[column] = list(map(convert_to_int, df[column]))
    return df,mapping


data_res = pd.read_csv('res.csv') 
data_com = pd.read_csv('com.csv')
data_all = data_res.append(data_com)

length  = len(data_all)
val_num = length/5;
data_num = length - val_num;

complete_data,mapping = convert_string_data_to_int(data_all)
complete_data['Year'] = complete_data.Year_Month.apply(lambda x: int(x/100))
complete_data['Month'] = complete_data.Year_Month.apply(lambda x:int( x%100))
complete_data['Date'] = pd.to_datetime(complete_data['Year_Month']*100+1, format='%Y%m%d').dt.strftime('%m/%d/%Y')

#drop row 201715 and 204406
complete_data = complete_data[complete_data.Year_Month != 201705]
complete_data = complete_data[complete_data.Year_Month != 204406]

complete_data_analysis(complete_data)
plotByPostalCode(complete_data)
plotByCustomerClass(complete_data, mapping)
```
